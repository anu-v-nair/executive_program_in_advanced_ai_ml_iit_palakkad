```python
class BankAccountManager:
  def __init__(self,account_number,balance=0):
    self.account_number = str(account_number)
    self.balance = balance

  def mask_account(self):
    return 'X'*(len(self.account_number) - 4) + self.account_number[-4:]

  def deposit_to_account(self,deposit_amount):
    if deposit_amount <= 0:
      return "Amount should be greater than zero"
    self.balance += deposit_amount
    return f"₹{deposit_amount} successfully deposited to {self.mask_account()}. Available Balance: ₹{self.balance:,.2f}"

  def withdraw_from_account(self,withdrawal_amount):
    if withdrawal_amount <= 0:
      return "Amount should be greater than zero"
    elif withdrawal_amount > self.balance:
      return f"Insufficient balance in {self.mask_account()}. Available Balance: ₹{self.balance:,.2f}"
    else:
      self.balance -= withdrawal_amount
      return f"₹{withdrawal_amount} successfully withdrawn from {self.mask_account()}. Available Balance: ₹{self.balance:,.2f}"

  def check_balance(self):
      return f"Balance Check successful. Available balance in {self.mask_account()} is ₹{self.balance:,.2f}"

account_1 = BankAccountManager(5874966524556)
account_2 = BankAccountManager(5874566221122,5000)
```

```python
account_1.check_balance()
```

```python
account_2.check_balance()
```

```python
withdrawal_amount = float(input("Please enter the amount to withdraw: ₹"))
account_1.withdraw_from_account(withdrawal_amount)
```

```python
import pandas as pd

class WeatherDataAnalyzer:
  def __init__(self,dataframe):
    self.df = dataframe
    self.df["Rain"] = self.df["Rain"].map({"Y":1,"N":0})

  def average_temperature(self):
    return self.df["Temperature"].mean()

  def average_humidity(self):
    return self.df["Humidity"].mean()

  def average_wind_speed(self):
    return self.df["Wind_Speed"].mean()

  def average_cloud_cover(self):
    return self.df["Cloud_Cover"].mean()

  def average_pressure(self):
    return self.df["Pressure"].mean()

  def rain_probability(self):
    return self.df["Rain"].mean() * 100

  def determine_weather(self):

    avg_temp = self.average_temperature()
    avg_humidity = self.average_humidity()
    avg_wind = self.average_wind_speed()
    avg_cloud = self.average_cloud_cover()
    avg_pressure = self.average_pressure()
    rain_chance = self.rain_probability()

    if rain_chance > 70 and avg_wind > 25:
      weather = "Stormy"

    elif rain_chance > 70:
      weather = "Heavy Rainfall"

    elif rain_chance > 40:
      weather = "Rainy"

    elif rain_chance > 20:
      weather = "Scattered Showers"

    elif avg_cloud > 80:
      weather = "Overcast"

    elif avg_cloud > 50:
      weather = "Partly Cloudy"

    elif avg_temp > 35 and avg_humidity > 70:
      weather = "Hot and Humid"

    elif avg_temp > 33:
      weather = "Very Hot"

    elif avg_wind > 25:
      weather = "Windy"

    elif avg_pressure < 1005:
      weather = "Low Pressure System"

    elif avg_temp < 20:
      weather = "Cool"

    else:
      weather = "Clear and Pleasant"

    return (
        f"Average Temperature: {avg_temp:.2f} °C\n"
        f"Average Humidity: {avg_humidity:.2f} %\n"
        f"Average Wind Speed: {avg_wind:.2f} km/h\n"
        f"Average Cloud Cover: {avg_cloud:.2f} %\n"
        f"Average Pressure: {avg_pressure:.2f} hPa\n"
        f"Rain Probability: {rain_chance:.2f} %\n"
        f"Weather Condition: {weather}\n"
    )

files = ["weather_forecast_data.csv","weather_forecast_data_2.csv","weather_forecast_data_3.csv","weather_forecast_data_4.csv"]
for file in files:
  df = pd.read_csv(file)
  weather = WeatherDataAnalyzer(df)
  print(weather.determine_weather())
```
