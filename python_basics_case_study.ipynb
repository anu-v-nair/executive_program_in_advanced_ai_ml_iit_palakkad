```python
class BankAccountManager:
  def __init__(self,balance=0):
    self.balance = balance

  def deposit_to_account(self):
    deposit_amount = float(input("Please enter deposit amount: "))
    self.balance += deposit_amount
    return f"Deposit successful. Account Balance: {self.balance}"

  def withdraw_from_account(self):
    withdraw_amount = float(input("Please enter withdrawal amount: "))
    if withdraw_amount > self.balance:
      return f"Insufficient funds. Current Balance: {self.balance}"
    else:
      self.balance -= withdraw_amount
      return f"Withdrawal successful. Account Balance: {self.balance}"

  def check_balance(self):
      return f"Current balance: {self.balance}"

account = BankAccountManager()
```

```python
account.deposit_to_account()
```

```python
account.withdraw_from_account()
```

```python
account.check_balance()
```
