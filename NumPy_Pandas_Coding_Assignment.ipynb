```python
import numpy as np
import pandas as pd
```

## Q1. Employee Salary Adjustment

```python
salaries = np.array([45000, 52000, 61000, 48000, 75000])

# 1. Increase all salaries by 10%.
salary_increase = salaries*1.10
print(f'Increased Salary: {salary_increase}')

# 2. Find the average salary.
average_salary = np.mean(salaries)
print(f'Average Salary: {average_salary:,.2f}')

# 3. Find the maximum and minimum salary.
max_salary = np.max(salaries)
min_salary = np.min(salaries)
print(f'Maximum Salary: {max_salary:,.2f} and Minimum Salary: {min_salary:,.2f}')

# 4. Print salaries greater than 60000.
filtered_salaries = salaries[salaries > 60000]
print(f'Salaries greater than 60,000: {filtered_salaries}')

```

# Q2. Sales Data Analysis

```python
sales = np.array([120,150,180,200,170,160,210,220,190,175])

# 1. Compute total sales.
total_sales = np.sum(sales)
print(f'Total Sales: {total_sales:,}')

# 2. Compute mean and standard deviation.
mean_sales = np.mean(sales)
sales_standard_deviation = np.std(sales)
print(f'Mean Sales: {mean_sales:,}\nStandard Deviation: {sales_standard_deviation:,}')

# 3. Identify days where sales exceeded the average.
days = np.where(sales > mean_sales)
print(f'Days where sales exceeded the average: {days}')

# 4. Normalize the sales using Min-Max scaling
normalized_sales = (sales - np.min(sales)) / (np.max(sales) - np.min(sales))
print(f'Normalized Sales: {normalized_sales}')
```

# Q3. Matrix Operations for Business Data

```python
sales_matrix = np.array([
[120,140,130,150],
[80,90,100,95],
[200,210,190,205]
])

# 1. Find total sales per product.
product_total = np.sum(sales_matrix,axis=1)
print(f'Total Sales Per Product: {product_total}')

# 2. Find total sales per region.
region_total = np.sum(sales_matrix,axis=0)
print(f'Total Sales Per Region: {region_total}')

# 3. Identify the best performing product.
best_product = np.argmax(product_total)
print(f'Best performing product: Product {best_product+1}')

# 4. Identify the region with highest sales
best_region = np.argmax(region_total)
print(f'Region with highest sales: Region {best_region+1}')
```

# Q4. Employee Dataset Exploration

```python
data = {
'Employee':['A','B','C','D','E','F'],
'Department':['HR','IT','IT','Finance','HR','Finance'],
'Salary':[50000,65000,70000,72000,52000,68000],
'Experience':[2,5,7,6,3,4]
}
# 1. Convert this dictionary into a DataFrame.
df = pd.DataFrame(data)

# 2. Display basic statistics of the dataset
print(f'Basic Statistics of the Dataset\n\n{df.describe()}\n\n')


# 3. Find the average salary per department.
dept_avg = df.groupby('Department')['Salary'].mean()
print(f'Average Salary per Department\n\n{dept_avg}\n\n')

# 4. Identify employees with experience greater than 4 years
employees = df[df['Experience'] > 4]
print(f'Employees with experience greater than 4 years\n\n{employees}\n\n')

# 5. Add a new column Bonus = 10% of salary
df['Bonus'] = df['Salary']*.10
print(f'DataFrame with Bonus column added\n\n',df)
```

# Q5. Missing Data Handling

```python
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}

df = pd.DataFrame(data)

# 1. Detect missing values.
print('Missing Values:\n\n',df.isnull())

# 2. Replace missing Age with mean age.
mean_age = df['Age'].mean().round(2)
df['Age'] = df['Age'].fillna(mean_age)

# 3. Replace missing Salary with median salary.
median_salary = df['Salary'].median()
df.fillna({'Salary': median_salary},inplace=True)

# 4. Remove rows with any remaining missing values.
df.dropna(inplace=True)
print('\nModified dataFrame\n\n',df)

```

# Q6. Sales Performance Analysis

```python
sales_data = {
'Region':['North','South','East','West','North','South','East','West'],
'Salesperson':['A','B','C','D','E','F','G','H'],
'Sales':[12000,15000,13000,17000,16000,14000,15500,16500]
}

df = pd.DataFrame(sales_data)

# 1. Find total sales per region.
region_total = df.groupby('Region')['Sales'].sum()
print(f'Total sales per region\n\n{region_total}\n\n')

# 2. Identify the top salesperson.
top_sps = df.loc[df['Sales'].idxmax()]
print(f'Top salesperson\n\n{top_sps}\n\n')

# 3. Compute average sales across regions.
region_avg = df.groupby('Region')['Sales'].mean()
print(f'Average sales across regions\n\n{region_avg}\n\n')


# 4. Sort the dataset by sales descending.
sorted_df = df.sort_values(by='Sales',ascending=False)
print('Sorted DataFrame\n\n',sorted_df)
```

# Q7. Customer Purchase Analysis

```python
data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

df = pd.DataFrame(data)


# 1. Find the average purchase amount by city.
city_avg = df.groupby('City')['PurchaseAmount'].mean()
print(f'Average purchase amount by city\n\n{city_avg}\n\n')

# 2. Identify customers with purchase above overall average.
avg_purchase = df['PurchaseAmount'].mean()
above_avg = df[df['PurchaseAmount'] > avg_purchase]
print(f'Customers Above Average\n\n {above_avg}\n\n')

# 3. Find the city generating the highest revenue.
city_revenue = df.groupby('City')['PurchaseAmount'].sum()
top_city = city_revenue.idxmax()
top_revenue = city_revenue.max()
print(f'City with Highest Revenue\n\n{top_city} : {top_revenue}\n\n')

# 4. Create a column Age_Group:
def group_age(age):
  if age < 30:
    return 'Young'
  elif age <= 50:
    return 'Adult'
  else:
    return 'Senior'


df['Age_Group'] = df['Age'].apply(group_age)
print(f'DataFrame with Age Group\n\n{df}\n\n')


# 5. Find average purchase per age group.
avg_purchase_per_age_group = df.groupby('Age_Group')['PurchaseAmount'].mean()
print(f'Average purchase per age group\n\n',avg_purchase_per_age_group)

```

# Q8. Time Series Sales Analysis

```python
dates = pd.date_range('2024-01-01', periods=12, freq='ME')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})

# 1. Convert Date to datetime index.
df.set_index('Date',inplace=True)

# 2. Compute monthly percentage growth.
df['Monthly_Growth(%)'] = df['Sales'].pct_change() * 100

# 3. Calculate 3-month rolling average sales.
df['Rolling_Average_Sales(3 months)'] = df['Sales'].rolling(window=3).mean().round(2)

print(f'Modified DataFrame\n\n{df}\n\n')

# 4. Identify the month with highest growth rate
max_growth_month = df['Monthly_Growth(%)'].idxmax().strftime('%B %Y')
max_growth_value = df['Monthly_Growth(%)'].max()
print(f'Month with highest growth rate\n\n{max_growth_month} : {max_growth_value}')
```
